```python
# Calculate win probabilities using Bernoulli/binomial formula
def win_probability(k, n):
    p = k / n
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

#Print probabilities
for team, wins in win_counts.items():
    prob = win_probability(wins, n)
    print(f"{team} won {wins} games — P(k={wins}, n=38): {prob:.8f}")
```
